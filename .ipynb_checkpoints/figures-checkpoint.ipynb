```python
from numpy import *
from numpy.random import normal
from scipy.linalg import eigh,circulant,toeplitz
from matplotlib.pyplot import *

def getLambda(n,D):
    
    X = normal(size=(n*n,n*n))
    
    k = circulant([-2,1]+(n-3)*[0]+[1])
    L = kron(k,eye(n)) + kron(eye(n),k)

    J = (X+X.T)/sqrt(8*n*n) + D*L
    return eigh(L,eigvals_only=True)
    
```

```python
Lambda = [ getLambda(150,0) for _ in xrange(1) ]
hist,bins=histogram(Lambda,bins=100,density=True)
bins = (bins[:-1]+bins[1:])/2
plot(bins,hist);
```
